```python
#11. Étudier les catégories avec le plus de contenu

#nous allons nous intéresser à la colonne listed_in. 
# La difficulté des colonnes listed_in, director et cast est qu’elles peuvent contenir des listes de valeurs
# Nous allons voir ensemble comment analyser des listes de valeurs en prenant comme exemple la colonne listed_in. 
# (adapter ce code ensuite pour les colonnes director et cast)

# aperçu de la colonne listed_in:
donnees['listed_in'].head()
```

```python
#On constate que chaque film/série peut appartenir à une ou plusieurs catégories. 
# Dans le cas où il appartient à plusieurs catégories, celles-ci sont séparées par une virgule et un espace 

# Nous allons voir une méthode Pandas join(), qui permet de joindre tous les éléments d’un tuple, 
# d’une liste ou même d’une série, séparés par un caractère que nous lui donnons en début de code.

# Pour illustrer ce propos, utilisons la méthode join() sur la colonne listed_in de notre dataframe 
# stocké dans la variable donnees et joignons l’ensemble des éléments de la colonne 
# avec le séparateur ", " (une virgule et un espace)
# On ignore les valeurs manquantes grâce à la méthode dropna()

categories=", ".join(donnees['listed_in'].dropna())
categories
```

```python
# Nous devons splitter, couper, ces catégories pour pouvoir les étudier
# (sinon Python considérera "Children &Family Movies, Comedies" comme une unique catégorie)
categories = categories.split(", ")

# Nous allons transformer cette liste en série, pour pouvoir ensuite utiliser la méthode value_counts() 
# (pour avoir le nombre d’occurrences dechaque catégorie dans le dataframe)
categories = pd.Series(categories)
categories.value_counts()
```

```python
#12. Afficher les directeurs qui ont produit le plus de films/séries disponibles sur Netflix
# adapter le code précédent pour la colonne director plutôt que la colonne listed_in, 
# afin d’afficher les dix directeurs de production les plus présents dans le jeu de données Netflix

director = pd.Series(",".join(donnees['director'].dropna()).split(",")) 
director.value_counts().head(10)
```

```python
#13. Voir si Jan Suter travaille souvent avec les mêmes acteurs
# afficher les cinq acteurs les plus présents dans les contenus produits par Jan Suter 
# en déduire si ce directeur travaille préférentiellement avec certains acteurs ou non.

# Garder uniquement les lignes du dataframe où les valeurs de la colonne director ne sont pas manquantes
# grâce à la méthode notna()
dataNa = donnees[donnees['director'].notna()]

# Utiliser la méthode str.contains() pour sélectionner uniquement les lignes dont la colonne director 
# contient la chaîne de caractères "Jan Suter" 
# ranger ce nouveau dataframe dans une variable nommée donnees_jan_suter
donnees_jan_suter = dataNa[dataNa['director'].str.contains('Jan Suter')]

# Adapter le code précédent, utilisé pour les colonnes listed_in et director
# afin d’afficher les cinq acteurs les plus présents dans le dataframe donnees_jan_suter.
actor_jan_suter = pd.Series(",".join(donnees_jan_suter['cast'].dropna()).split(","))
actor_jan_suter.value_counts().head(5)
```

```python
#14. Représenter les dix pays qui ont produit le plus de contenus disponibles sur Netflix, 
# avec le nombre de contenus par pays
# Sélectionner les dix pays ayant produit le plus de contenus grâce à la méthode value_counts() 
# etla méthode head() et stocker cette série dans une variable.
data_pays = pd.Series(", ".join(donnees['country'].dropna()).split(", ")) 
data_pays_10 = data_pays.value_counts().head(10)
data_pays_10
```

```python
#Créer un dataframe contenant uniquement les lignes pour ces dix pays
# i-e les lignes pour lesquelles la valeur dans la colonne country est comprise 
# dans les index de la série créée précédemment. 
# utiliser l’indexation booléenne, la méthode .isin() et l’attribut index
df_pays_10 = donnees[donnees['country'].isin(data_pays_10.index)]
df_pays_10
```

```python
# Tracer un countplot grâce à la fonction countplot() de Seaborn  
# en lui donnant le nouveau dataframe et la variable souhaitée : country.

import seaborn as sns
sns.countplot(x="country", data=df_pays_10)
```

```python
# pour améliorer la lisibilité du graphe : utiliser pyplot (plt)
import matplotlib.pyplot as plt

plt.figure(figsize= [15, 6])
sns.countplot(x="country", data=df_pays_10)
```

```python
# 15. Tracer un graphe à barres du nombre de films/séries parclassement de contenu (rating)
#tracer le graphe à barres en utilisant le dataframe
# d’origine et la colonne d’intérêt représentant le classement de contenu
plt.figure(figsize= [15, 6])

sns.countplot(x="rating", data=donnees, order=donnees['rating'].value_counts().keys())
```

```python
#16. Afficher l’évolution du nombre de films/séries disponibles sur Netflix au cours du temps

# écrire le code pour formater correctement la colonne date_added et créer une nouvelle colonne 
# dans le dataframe contenant uniquement l’année d’ajout du contenu sur Netflix.
donnees["year_added"] = pd.to_datetime(donnees.date_added).dt.year

# Grouper le dataframe sur les variables year_added et type, afin de tracer 
# l’évolution du nombre de contenus ajoutés par an grâce à la fonction pointplot() de Seaborn. 
# Séparer les courbes selon le type ducontenu, Movie ou TV Show
df_year_added_type= donnees.groupby(by=["year_added", "type"]).size().reset_index(name="count")

plt.figure(figsize= [15, 6])

sns.pointplot(x="year_added", y="count", hue="type", data=df_year_added_type)
```
